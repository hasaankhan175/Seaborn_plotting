```python
# set Style of the plot 
# Box plot 
sns.set_theme(style='darkgrid' , font_scale= 1.3 , font='Times New Roman')

plt.figure(figsize=(8,  4)) # it is the size of the plot

palette2 = {'First' : 'red' , 'Second' : 'orange'  , 'Third' : 'yellow'}


sns.boxplot(data = df , x = 'sex' , y = 'age', hue = 'class' , palette= palette2 ) # change the scatter points color based on hue 


# you can add labels here 
plt.xlabel('Fare') # x-axis
plt.ylabel('Age') # y-axis
plt.title('Scatter plot of age vs fare') # title


# change legend position and title 
#'mid' is not a valid value for loc; supported values are 'best', 'upper right', 'upper left', 'lower left', 'lower right', 'right', 'center left', 
# 'center right', 'lower center', 'upper center', 'center'
plt.legend(loc = 'right' , title = 'Class')
```

```python
# set Style of the plot 
# Box plot 
sns.set_theme(style='darkgrid' , font_scale= 1.3 , font='Times New Roman')

plt.figure(figsize=(8,  4)) # it is the size of the plot

palette2 = {'First' : 'red' , 'Second' : 'orange'  , 'Third' : 'yellow'}

# we convert the box plot into boxen plot
sns.boxenplot(data = df , x = 'sex' , y = 'age', hue = 'class' , palette= palette2 ) # change the scatter points color based on hue 


# you can add labels here 
plt.xlabel('Fare') # x-axis
plt.ylabel('Age') # y-axis
plt.title('Scatter plot of age vs fare') # title


# change legend position and title 
#'mid' is not a valid value for loc; supported values are 'best', 'upper right', 'upper left', 'lower left', 'lower right', 'right', 'center left', 
# 'center right', 'lower center', 'upper center', 'center'
plt.legend(loc = 'right' , title = 'Class')
 
# change the rotation of the x stick labels 
plt.xticks(rotation = 45)
```

## ***BAR PLOT***

```python
# set Style of the plot 
# Box plot 
sns.set_theme(style='darkgrid' , font_scale= 1.3 , font='Times New Roman')

plt.figure(figsize=(8,  4)) # it is the size of the plot

palette2 = {'First' : 'red' , 'Second' : 'orange'  , 'Third' : 'yellow'}

# we convert the box plot into boxen plot
sns.barplot(data = df , x = 'sex' , y = 'age', hue = 'class' , palette= palette2  , capsize = 0.1) # change the scatter points color based on hue 


# you can add labels here 
plt.xlabel('Fare') # x-axis
plt.ylabel('Age') # y-axis
plt.title('Scatter plot of age vs fare') # title


# change legend position and title 
#'mid' is not a valid value for loc; supported values are 'best', 'upper right', 'upper left', 'lower left', 'lower right', 'right', 'center left', 
# 'center right', 'lower center', 'upper center', 'center'
plt.legend(loc = 'right' , title = 'Class')
 
# change the rotation of the x stick labels 
plt.xticks(rotation = 45)
```
